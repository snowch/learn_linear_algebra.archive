### Graphics

Plot our vector.

```python
a.plot(figsize=4)
```

```python
a = vector([1,2])
b = vector([1,-2])

plots = a.plot() + b.plot()
show(plots, figsize=4)
```

##### Linear Combination

```python
a = vector([1,2])
b = vector([1,-2])

a2 = 2*a

plots = a.plot() + b.plot() + a2.plot()
show(plots, figsize=4)
```

```python
a = vector([1,2])
b = vector([1,0])

c = a + b

plots = a.plot() + b.plot() + c.plot(color='red')
show(plots, figsize=4)
```

### n-Dim Vectors
