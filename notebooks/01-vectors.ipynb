### Linear Combinations

> The linear combination of the vectors  $𝐯_1,𝐯_2,…,𝐯_𝑛$ with scalars  $𝑐_1,𝑐_2,…,𝑐_𝑛$ is the vector
> 
> $𝑐_1𝐯_1+𝑐_2𝐯_2+…+𝑐_𝑛𝐯_𝑛.$
> 
> The scalars $𝑐_1,𝑐_2,…,𝑐_𝑛$ are called the weights of the linear combination.

#### Scalar multiplication

```python
v1 = vector([1,2])
c1 = 2

v1xc1 = c1*v1

plots = v1.plot(color='blue') + v1xc1.plot(color='lightblue', linestyle=':') 
show(plots, figsize=4)
```

#### Vector addition

```python
v1 = vector([1,2])
v2 = vector([1,0])

v1_v2 = v1 + v2

plots = v1.plot(color='blue') + v2.plot(color='blue') + v1_v2.plot(color='red')
show(plots, figsize=4)
```
